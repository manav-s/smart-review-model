```python
def get_adjectives(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "JJ"]

df['adjectives'] = df['Review'].apply(get_adjectives)
```

```python
five_star = df[df['Stars'] == 5]
one_star = df[df['Stars'] == 1]
```

```python
five_star['adjectives'] = five_star.adjectives.apply(' '.join)
five_star
```

```python
good_words = Counter(" ".join(five_star["adjectives"]).split()).most_common()
```

```python
one_star['adjectives'] = one_star.adjectives.apply(' '.join)
one_star
```

```python
Counter(" ".join(one_star["adjectives"]).split()).most_common()
```

```python
# top 10 good words
keywords = good_words[0:20] 

for index,tup in enumerate(keywords):
    keywords[index] = tup[0]
    
keywords
```

```python
for word in keywords:
    column_name = 'has_' + word
    df[column_name] = df.Review.apply(lambda x: word in x)
```

```python
df
```

```python
# convert the True and False to 1 and 0
df = df * 1
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### RandomForestRegressor

```python
X   = df[['has_great', 'has_good', 'has_nice', 'has_other', 'has_delicious', 'has_friendly', 'has_little', 'has_first', 'has_awesome', 'has_favorite', 'has_many', 'has_few', 'has_much', 'has_sure', 'has_next', 'has_happy', 'has_perfect']]
y   = df['Stars']            

X_train, X_test, y_train, y_test = train_test_split(X,              #the input features
                                                    y,              #the label
                                                    test_size=0.3,  #set aside 30% of the data as the test set
                                                    random_state=7 #reproduce the result
                                                   )
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=7)
rf.fit(X_train, y_train)
```

```python
y_pred   = rf.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print('The mse of the model is: {}'.format(mse))
```

### SVC

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
X   = df[['has_great', 'has_good', 'has_nice', 'has_other', 'has_delicious', 'has_friendly', 'has_little', 'has_first', 'has_awesome', 'has_favorite', 'has_many', 'has_few', 'has_much', 'has_sure', 'has_next', 'has_happy', 'has_perfect']]
y   = df['Stars']            

X_train, X_test, y_train, y_test = train_test_split(X,              #the input features
                                                    y,              #the label
                                                    test_size=0.3,  #set aside 30% of the data as the test set
                                                    random_state=7, #reproduce the results
                                                    stratify=y)
```

```python
scaler = StandardScaler()
scaler.fit(X_train[['has_great', 'has_good', 'has_nice', 'has_other', 'has_delicious', 'has_friendly', 'has_little', 'has_first', 'has_awesome', 'has_favorite', 'has_many', 'has_few', 'has_much', 'has_sure', 'has_next', 'has_happy', 'has_perfect']])

X_train[['has_great', 'has_good', 'has_nice', 'has_other', 'has_delicious', 'has_friendly', 'has_little', 'has_first', 'has_awesome', 'has_favorite', 'has_many', 'has_few', 'has_much', 'has_sure', 'has_next', 'has_happy', 'has_perfect']] = scaler.transform(X_train[['has_great', 'has_good', 'has_nice', 'has_other', 'has_delicious', 'has_friendly', 'has_little', 'has_first', 'has_awesome', 'has_favorite', 'has_many', 'has_few', 'has_much', 'has_sure', 'has_next', 'has_happy', 'has_perfect']])
```

```python
X_train.head(10)
```

```python
params = {'C':np.arange(1,3),
         'gamma':np.arange(1,3)}

svm = SVC()
svm_grid = GridSearchCV(svm, params, cv=5)
svm_grid.fit(X_train, y_train)

print(svm_grid.best_params_)
print(svm_grid.best_score_)
```

```python
best_model = SVC(**svm_grid.best_params_).fit(X_train,y_train)
```
