```python
def get_adjectives(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "JJ"]

df['adjectives'] = df['Review'].apply(get_adjectives)
```

```python
five_star = df[df['Stars'] == 5]
one_star = df[df['Stars'] == 1]
```

```python
five_star['adjectives'] = five_star.adjectives.apply(' '.join)
five_star
```
